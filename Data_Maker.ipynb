```python
import scanpy as sc

adata = sc.read_h5ad("mouse_brain_vizgen.h5ad")
```

```python
import celltypist

# Normalize counts
sc.pp.normalize_total(adata, target_sum=1e4)

# Log transform
sc.pp.log1p(adata)

# Run CellTypist
predictions = celltypist.annotate(
    adata,
    model='Developing_Mouse_Brain.pkl',
    majority_voting=True
)

# Add labels into adata
adata.obs["Cell Type"] = predictions.predicted_labels

print(adata.obs["Cell Type"].value_counts())
```

```python
adata.obs["Cell Type"] = predictions.predicted_labels["majority_voting"]
```

```python
import scanpy as sc

sc.pl.scatter(
    adata,
    x="min_x",
    y="min_y",
    color="Cell Type"
)
```

```python
sc.pp.pca(adata)
sc.pp.neighbors(adata)
sc.tl.umap(adata)

sc.pl.umap(
    adata,
    color="Cell Type"
)
```

```python
sc.tl.leiden(adata, resolution=2.0)
```

```python
adata.obs["spatial_domain"] = adata.obs["leiden"]
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,10))

plt.scatter(
    adata.obsm["spatial"][:, 0],
    adata.obsm["spatial"][:, 1],
    c=adata.obs["Cell Type"].astype("category").cat.codes,
    s=1,
    cmap="tab20"
)

plt.gca().invert_yaxis()
plt.title("Cell Type (MERFISH)")
plt.legend("Cell Type")
plt.show()
```

```python
cell_type = "Neural tube: Anterior"

subset = adata[adata.obs["Cell Type"] == cell_type]

plt.figure(figsize=(10,10))
plt.scatter(
    subset.obsm["spatial"][:, 0],
    subset.obsm["spatial"][:, 1],
    s=1
)

plt.gca().invert_yaxis()
plt.title(cell_type)
plt.show()
```
